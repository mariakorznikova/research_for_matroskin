```python
import pandas as pd
import requests
import time
```

```python
# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã –∑–∞–ø—Ä–æ—Å–∞ –∫ VK API
params = {
    'access_token' : '–∑–¥–µ—Å—å –±—ã–ª —Ç–æ–∫–µ–Ω',
    'v' : '5.199',
    'domain' : 'matroskinperm',
    'extended' : '1',
    'offset': 0
}
```

```python
# –û—Ç–ø—Ä–∞–≤–∫–∞ –∑–∞–ø—Ä–æ—Å–∞ –∫ API VK —Å —É–∫–∞–∑–∞–Ω–Ω—ã–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏
url = 'https://api.vk.ru/method/wall.get' 
response = requests.get(url, params=params) 
response = response.json() 
response
```

```python
total_count = response['response']['count']
print(f'–í—Å–µ–≥–æ –ø–æ—Å—Ç–æ–≤: {total_count}')
```

```python
pd.json_normalize(response) # –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –≤–ª–æ–∂–µ–Ω–Ω–æ–≥–æ JSON-–æ–±—ä–µ–∫—Ç–∞ response –≤ –ø–ª–æ—Å–∫—É—é —Ç–∞–±–ª–∏—Ü—É (DataFrame)
```

```python
response.keys() # –ü—Ä–æ—Å–º–æ—Ç—Ä –∫–ª—é—á–µ–π –≤–µ—Ä—Ö–Ω–µ–≥–æ —É—Ä–æ–≤–Ω—è –≤ —Å–ª–æ–≤–∞—Ä–µ response
```

```python
response['response'].keys() # –ü—Ä–æ—Å–º–æ—Ç—Ä –∫–ª—é—á–µ–π –≤–Ω—É—Ç—Ä–∏ –≤–ª–æ–∂–µ–Ω–Ω–æ–≥–æ —Å–ª–æ–≤–∞—Ä—è —Å –∫–ª—é—á–æ–º 'response
```

```python
 df = pd.json_normalize(response['response']['items']) # –°–æ–∑–¥–∞–Ω–∏–µ DataFrame –∏–∑ —Å–ø–∏—Å–∫–∞ –ø–æ—Å—Ç–æ–≤, —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω–Ω–æ–≥–æ –≤ response['response']['items']
 df
```

```python
all_posts = [] # –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è —Å–ø–∏—Å–∫–∞ –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è –≤—Å–µ—Ö –∑–∞–≥—Ä—É–∂–µ–Ω–Ω—ã—Ö –ø–æ—Å—Ç–æ–≤
count = 100  # –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ –∑–∞–ø–∏—Å–µ–π –∑–∞ –æ–¥–∏–Ω –∑–∞–ø—Ä–æ—Å (–º–∞–∫—Å–∏–º—É–º 100)
offset = 0 # —Å–º–µ—â–µ–Ω–∏–µ –¥–ª—è –ø–æ—Å—Ç—Ä–∞–Ω–∏—á–Ω–æ–≥–æ –ø–æ–ª—É—á–µ–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö, –Ω–∞—á–∏–Ω–∞–µ–º —Å 0
# –¶–∏–∫–ª –ø–æ–∫–∞ —Å–º–µ—â–µ–Ω–∏–µ –º–µ–Ω—å—à–µ –æ–±—â–µ–≥–æ —á–∏—Å–ª–∞ –ø–æ—Å—Ç–æ–≤
while offset < total_count:
    print(f'–ó–∞–≥—Ä—É–∂–∞—é –ø–æ—Å—Ç—ã —Å {offset} –ø–æ {offset + count}')
    params.update({'count': count, 'offset': offset})
    response = requests.get(url, params=params).json()

    # –ï—Å–ª–∏ –æ—Ç–≤–µ—Ç —Å–æ–¥–µ—Ä–∂–∏—Ç –æ—à–∏–±–∫—É ‚Äî –ø—Ä–µ—Ä–≤—ë–º —Ü–∏–∫–ª
    if 'error' in response:
        print("–û—à–∏–±–∫–∞ API:", response['error'])
        break

    items = response['response']['items']
    all_posts.extend(items)

    offset += count

    time.sleep(0.34)  # –ª–∏–º–∏—Ç—ã VK API ‚Äî –Ω–µ –±–æ–ª–µ–µ 3 –∑–∞–ø—Ä–æ—Å–æ–≤ –≤ —Å–µ–∫—É–Ω–¥—É (–º–æ–∂–Ω–æ —É–≤–µ–ª–∏—á–∏—Ç—å –∑–∞–¥–µ—Ä–∂–∫—É)

# –°–æ–∑–¥–∞—ë–º DataFrame –∏–∑ –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –ø–æ—Å—Ç–æ–≤
df = pd.json_normalize(all_posts)
print(f'–°–∫–∞—á–∞–Ω–æ –ø–æ—Å—Ç–æ–≤: {len(df)}')
```

```python
import pandas as pd
import requests
import time
```

```python
# Параметры запроса к VK API
params = {
    'access_token' : 'здесь был токен',
    'v' : '5.199',
    'domain' : 'matroskinperm',
    'extended' : '1',
    'offset': 0
}
```

```python
# Отправка запроса к API VK с указанными параметрами
url = 'https://api.vk.ru/method/wall.get' 
response = requests.get(url, params=params) 
response = response.json() 
response
```

```python
total_count = response['response']['count']
print(f'Всего постов: {total_count}')
```

```python
pd.json_normalize(response) # Преобразование вложенного JSON-объекта response в плоскую таблицу (DataFrame)
```

```python
response.keys() # Просмотр ключей верхнего уровня в словаре response
```

```python
response['response'].keys() # Просмотр ключей внутри вложенного словаря с ключом 'response
```

```python
 df = pd.json_normalize(response['response']['items']) # Создание DataFrame из списка постов, расположенного в response['response']['items']
 df
```

```python
all_posts = [] # инициализация списка для хранения всех загруженных постов
count = 100  # максимальное число записей за один запрос (максимум 100)
offset = 0 # смещение для постраничного получения данных, начинаем с 0
# Цикл пока смещение меньше общего числа постов
while offset < total_count:
    print(f'Загружаю посты с {offset} по {offset + count}')
    params.update({'count': count, 'offset': offset})
    response = requests.get(url, params=params).json()

    # Если ответ содержит ошибку — прервём цикл
    if 'error' in response:
        print("Ошибка API:", response['error'])
        break

    items = response['response']['items']
    all_posts.extend(items)

    offset += count

    time.sleep(0.34)  # лимиты VK API — не более 3 запросов в секунду (можно увеличить задержку)

# Создаём DataFrame из полученных постов
df = pd.json_normalize(all_posts)
print(f'Скачано постов: {len(df)}')
```
